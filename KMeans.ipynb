```python
#Menentukan Cluster  Terbaik (Elbow Method)
#Hitung Within Cluster sum of Squared error (WCSS) untuk tiap kluster yang berbeda
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmModel=KMeans(n_clusters=i,n_init=10)
    kmModel.fit(df)
    wcss.append(kmModel.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss)
plt.plot(range(1,11), wcss, linewidth=2, color='red', marker='8')
plt.xlabel('Jumlah Cluster')
plt.xticks(np.arange(1,11,1))
plt.ylabel('WCSS')
plt.show()
```

```python
# Model Dengan 5 Cluster
kmModelfix=KMeans(n_clusters=2,n_init=10)
kmModelfix.fit(df)

#Memprediksi Label
hasil_prediksi = kmModelfix.predict(df)
print(hasil_prediksi)
```

```python
df['Label']=hasil_prediksi
df.head()
```

```python
df
```

```python
from sklearn.metrics import silhouette_score

# df adalah data yang digunakan untuk klastering
# hasil_prediksi adalah hasil klaster yang didapat dari model klastering
silhouette_avg = silhouette_score(df, hasil_prediksi)

print(f"Silhouette Score: {silhouette_avg}")
```
